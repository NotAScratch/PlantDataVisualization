# Bar Plot
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

## Components of a Bar Graph
1. Chart Title:Chart Title: It denotes the name of the bar chart. In this, we can write what the chart is representing.

2. Grid Lines: The vertical and horizontal lines in gray color is called grid lines.

3. Bars: A bar is corresponding to a value. It may be horizontal or vertical. The largest bar represents the largest value.

4. Axis Title: A bar graph has two titles one is vertical, and the other is horizontal. Both the axis is related to each other. We can write the axis title for easy understanding. Suppose, the vertical axis represents expenses. So, we can write Expenses (in rupees) on the vertical axis. The expenses may be of different types, so we can write types of expenses on the horizontal axis.

5. Labels: We can also categorize the horizontal axis title. For example, types of expenses can be categorized into medical, transport, office, etc.

6. Legends: A legend specifies what a bar is representing. It is also known as the key of a chart. Consider the following graph; if we write 2019 in place of Series 1, it means the blue bars in the graph represent the data of the year 2019.

7. Scale: The scale represents the vertical values. It may include rupees, population, size, etc.






```python
x= [2001,2002,2003,2005,2006]
y = [900,600,700,800,500]
plt.figure(figsize=(8, 6))
plt.bar(x, y)
plt.xlabel("Year")
plt.ylabel("Values")
plt.title("Bar Plot of Data")
plt.show()
```

# Pie Plot
A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data.

```python
#autopct is an argument used in the plot() function to display the percentage labels on the pie chart. The autopct parameter formats the percentage values shown on each slice of the pie chart.
data["Species"].value_counts().plot(kind="pie", autopct="%.2f%%")
```

# Scatter Plot
A scatter plot is a diagram where each value in the data set is represented by a dot on a two-dimensional coordinate system. It is used to visually explore the relationship between two continuous variables.

```python

data.plot(kind ="scatter",
          x ='SepalLengthCm',
          y ='PetalLengthCm')
plt.grid()
```

```python
#Scatter plot
#Defining list of colors and different species
colors= ['red','orange','blue']
species= ['Iris-virginica','Iris-versicolor','Iris-setosa']
```

```python
#Scatter plot using Sepal length and Sepal width
for i in range(3):
    x=data[data['Species']==species[i]]
    plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'],c=colors[i],label=species[i])
#To give the level
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

#To display the different classes in clors
plt.legend()

    		
```

```python
#Scatter plot using petal length and width 
for i in range(3):
    x=data[data['Species']==species[i]]
    plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'],c=colors[i],label=species[i])
#To give the lave
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")

#To display the different classes
plt.legend()
```

```python
#Scatter plot using sepal and petal length 
for i in range(3):
    x=data[data['Species']==species[i]]
    plt.scatter(x['SepalLengthCm'],x['PetalLengthCm'],c=colors[i],label=species[i])
#To give the lave
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

#To display the different classes
plt.legend()
```

```python
#Scatter plot using sepal and petal  width wise
for i in range(3):
    x=data[data['Species']==species[i]]
    plt.scatter(x['SepalWidthCm'],x['PetalWidthCm'],c=colors[i],label=species[i])
#To give the lave
plt.xlabel("Sepal width")
plt.ylabel("Petal Width")

#To display the different classes
plt.legend()
```

# Coorelation matrix
A Correlation matrix is a matrix which shows the correlation between variables. Each cell in the matrix shows in corellation between 2 variables . The value is in the range of -1 to 1. If two variables have high correlation we can neglate one variable frome those two.

```python
#Display correlation in tabular form
data.corr()
```

```python
#using heat map
corr=data.corr()
plt.figure(figsize=(8, 6))
#The annot=True parameter adds numerical annotations to each cell in the heatmap, displaying the correlation coefficients. 
#
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
